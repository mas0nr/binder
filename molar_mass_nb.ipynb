```python
import pandas as pd
```

```python
def breakup():
    indices = []
    molecule = input("molecule here")
    for i in range(len(molecule)):
        if molecule[i].isupper():
            indices.append(i)
    #indices.append(len(molecule))
    #print(indices)
    splitstring = []
    for i in range(len(indices)):
        try:
            splitstring.append(molecule[indices[i]:indices[i+1]])
        except IndexError:
            splitstring.append(molecule[indices[i]:])
    return splitstring
```

```python
def calcmm(df,atoms):
    total = 0
    for m in atoms:
        if m.isalnum():
            abbr = "".join(c for c in m if c.isalpha())
            print(f"{abbr} - {df[abbr]} g/mol")
            count = int("".join(c for c in m if c.isnumeric()))
            mass = df[abbr]
            total += count*mass 
        else:
            total += df[m]
    print(f"{total} g/mol")
elements_df = pd.read_csv("ptableall.csv").set_index("Element")["Atomic Mass"].to_dict() # converts CSV to dataframe and then dataframe to dictionary with the first column as keys and the second as values
```

```python
splitstring = breakup()
calcmm(elements_df,splitstring)
```
